```python
import random
from game import Game, Move, Player
from QPlayer import QPlayer, Environment, train, QLearningAgent

import numpy as np
import pickle 
import os
import zipfile


# Seed value
# Apparently you may use different seed values at each stage
seed_value= 2

# 1. Set the `PYTHONHASHSEED` environment variable at a fixed value
os.environ['PYTHONHASHSEED']=str(seed_value)

# 2. Set the `python` built-in pseudo-random generator at a fixed value
random.seed(seed_value)

# 3. Set the `numpy` pseudo-random generator at a fixed value
np.random.seed(seed_value)
```

```python
class RandomPlayer(Player):
    def __init__(self) -> None:
        super().__init__()

    def make_move(self, game: 'Game') -> tuple[tuple[int, int], Move]:
        from_pos = (random.randint(0, 4), random.randint(0, 4))
        move = random.choice([Move.TOP, Move.BOTTOM, Move.LEFT, Move.RIGHT])
        return from_pos, move
```

```python
player1 = QPlayer()
player1.agent = QLearningAgent(epsilon=None)
with open('qtable0.pkl', 'rb') as f:
    player1.agent.q_table  = pickle.load(f)
```

```python
counter = 0

for i in range(100):
    g = Game()
    g.print()

    env = Environment()
    env.game = g
    player1.env = env

    player2 = RandomPlayer()
    winner = g.play(player1, player2)
    if winner == 0: ##winner ==1 in case we are the second player
        counter+= 1
    g.print()

    print(f'-----------------{i}-winner_id: {winner}-----------------')

print(counter)
print(player1.agent.usefullness / player1.agent.total)
```
