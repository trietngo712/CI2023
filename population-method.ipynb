## Evolution 

- Accumulation of tiny variations

- Evolution is a sequence of two contributions:
    - Variations (Random)
    - Selection (Deterministic)

- In evolution, changes are not designed but evalutated

- Evolution is not an optimization process
- Evolution does not have a goal
- Evolution does not favor strength
- Evolution does not favor intelligence

### Essential of Evolutionary Algorithm

- Candidate solution -> Individual
- Set of candidate solutions -> Population
- Ability to solve the problem -> Fitness
- Sequence of steps -> Generations

- Individual -> Encode a potential solution for the problem
- Parent selection
- Genetic Operators
- Evaluation -> Survival Selection

## Representation

- Fitness: how well the candidate solution is able to solve the target problem
- Genotype: the internal representation of the individual, what is directly manipulated by genetic operators
- Phenotype: the candidate solution that is encoded in the genotype
    - the intermediate form in which the genotype needs to be transformed in to for evaluating fitness
    - if genotype can be directly evaluated: genotype and phenotype coincide

- Importance of mappings:
     - small differences shoul bring small differences in the fitness (locality) -> strong causality

- Genetic operators
    - mutation
    - recombination (crossover)
    - in general its better to have both mutation and recombination operators
        - mutation : small tweaks, exploitation
        - recombination: big changes, exploration
    - mutation only could be fine
    - recombination only would not work
- Importance of encoding
    - genetic operators should make sense on phenotype level

- Binary representation
    - mutation : bitflip
    - recombination : 1-cut crossover
        - positional bias:
            - more likely to keep together genes that are near
            - can never keep together genes from opposite ends of strings

        - can be exploited if we know the structure
    - recombination : n - cut crossover

    - recombination: uniform crossover
        - no positional bias
- Integer Representation
    - same caveats as binary representation
    - Categorical vs Discrete

    - mutation: random resetting
    - mutation: creep mutation
    - recombination : same as binary representation

- Permutation representation

    the information is stored in the absolute position, order, sequence
    - mutation: swap mutation
    - mutation: scramble mutation
    - mutation: insert mutation
    - mutation: inversion mutation

    -recombination: Cycle crossover
    -recombination: partially mapped crossover
    -recombination: inver over
